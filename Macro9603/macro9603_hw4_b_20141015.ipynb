### (i) Solve for the steady state capital stock

Write a program to numerically solve the difference equation from part (b), both for a steady state as well for $k_{t+1}$ given any $k_t$, for general values of the parameters $\beta, \gamma, A, \alpha, l_1, l_2, \tau$. 

The first-order difference equation that governs the dynamics of equilibrium capital stock, $g(k_t, k_{t+1};\theta)$, where $\theta = (\beta, \gamma, A, \alpha, l_1, l_2, \tau)$ is the vector of parameters. 

$$
\newcommand{\oa}[1]{\left(#1\right)}
\newcommand{\ob}[1]{\left[#1\right]}
\newcommand{\oc}[1]{\left\{#1\right\}}
$$

In this economy $g(k_t, k_{t+1};\theta)$ is given by
$$
k_{t+1} -
\frac{
(1-\alpha)A\oa{\frac{k_t}{l_1+l_2}}^\alpha l_1(1-\tau)\beta^{1/\sigma}
\times
\ob{1 + \alpha A \oa{\frac{l_1 + l_2}{k_{t+1}} }^{1-\alpha} - \delta }^{1/\sigma}
- (1-\alpha)A \oa{\frac{k_{t+1}}{l_1 + l_2} }^{\alpha}(l_2 + \tau l_1) 
}{
\beta^{1/\sigma}\oa{1 + \alpha A \oa{\frac{l_1 + l_2}{ k_{t+1}}}^{1-\alpha}  -\delta}^{1/\sigma}
+ \oa{1 + \alpha A \oa{\frac{l_1 + l_2}{ k_{t+1}}}^{1-\alpha}  -\delta}
} = 0
$$




In steady state $k_{t+1} = k_t = k$

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
#Parameters 
beta = 1
gamma = 1
A = 25
alpha = 1/3
delta = 1/2
l1 = 1
l2 = 0 
tau = 0
parameter = np.array([beta, gamma, A, alpha, delta, l1, l2, tau])
sigma = 1
k_min = 0
k_max = 30
```

```python
# First-order difference equation of the equilibrium capital stock

#Note that when gamma==1, we get the same difference equation

def g(k,k1,theta,sigma):
    beta, gamma, A, alpha, delta, l1, l2, tau = theta
    r=alpha*A*((l1+l2)/k1)**(1-alpha) #r_{t+1}
    w=(1-alpha)*A*(k/(l1+l2))**alpha #w_t
    w1=(1-alpha)*A*(k1/(l1+l2))**alpha #w_t+1
    A = beta**(1/sigma)*(1+r-delta)**(1/sigma)
    B = 1+r-delta
    D = w1*l2 + tau*w1*l1
    E = w*l1*(1-tau)
    s = (A*E - D)/(A+B)
    return k1-s

def gss(k):
    kss = g(k,k,parameter,sigma)
    return kss

def gt(k):
    return g(k0,k,parameter,sigma)
```

```python
# Bisection method

def bisect(f, a, b, tol=10e-8):
    lower, upper = a, b
    global point
    point = []
    if upper - lower < tol:
        point1 = 0.5 * (upper + lower)  
        point.append(point1)
        return point
    else:
        middle = 0.5 * (upper + lower)
        if f(middle) > 0:   
            bisect(f, lower, middle)
        else:               
            bisect(f, middle, upper)
        point.append(middle)
        point = point[:1]
    return  point[0]
```

```python
#Steady state and capital stock grid 

kss = bisect(gss,k_min,k_max)
k_grid = np.linspace(0, 1.1*kss, 50) #Grid of capital

print 'Steady state capital stock:',kss
```

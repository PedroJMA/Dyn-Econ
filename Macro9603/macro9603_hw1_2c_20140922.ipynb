Suppose that $\beta_2 = 0.75$ and $\beta_1 = \gamma \beta_2$. Consider two different cases, $\gamma = 0.5$ and $\gamma = 0.95$. Plot the equilibrium consumption of the two consumers and equilibrium prices for the first 100 periods for these two cases. Explain the intuition for why the time paths of consumption vary with $\gamma$. 

Equilibrium allocations of consumption are given by: 
$$
C^1_t = 
\frac{ 6(1-\beta_1)\beta_1^t }{ 2(1-\beta_1)\beta_1^t + (1-\beta_2)\beta_2^t}
$$

$$
C^2_t = 
\frac{ 3(1-\beta_2)\beta_2^t }{ 2(1-\beta_1)\beta_1^t + (1-\beta_2)\beta_2^t}
$$

### Case 1: $\gamma= 0.5$

```python
%matplotlib inline
import matplotlib.pyplot as plt

time = 100
gamma = 0.5
beta2 = 0.75
beta1 = gamma*beta2

def c1(t, coef): 
    b1, b2 = coef
    return ( 6*(1-b1)*b1**t ) / ( 2*(1-b1)*b1**t + (1-b2)*b2**t)

def c2(t, coef): 
    b1, b2 = coef
    return ( 3*(1-b2)*b2**t ) / ( 2*(1-b1)*b1**t + (1-b2)*b2**t)

def price(t, coef):
    b1, b2 = coef
    return ( 2*(1-b1)*b1**t + (1-b2)*b2**t ) / ( 2*(1-b1) + (1-b2) )

c1_value = [c1(t,(beta1,beta2)) for t in range(time)]
c2_value = [c2(t,(beta1,beta2)) for t in range(time)]
price_1 = [price(t,(beta1,beta2)) for t in range(time)]

plt.plot(c1_value, '-b' , label='Consumption allocation 1', linewidth=2 )
plt.plot(c2_value, 'b--', label='Consumption allocation 2', linewidth=2 )
plt.ylabel(' ')
plt.xlabel('Time', fontsize=18)
plt.title('Case 1: gamma=0.5', fontsize=20)
plt.legend(loc='center right')
plt.show()
```

Note that in this case $\beta_1=0.375$ and $\beta_2=0.75$

### Case 2: $\gamma= 0.95$

```python
gamma = 0.95
beta1 = gamma*beta2

c1_value = [c1(t,(beta1,beta2)) for t in range(time)]
c2_value = [c2(t,(beta1,beta2)) for t in range(time)]
price_2 = [price(t,(beta1,beta2)) for t in range(time)]

plt.plot(c1_value, '-b' , label='Consumption allocation 1', linewidth=2 )
plt.plot(c2_value, 'b--', label='Consumption allocation 2', linewidth=2 )
plt.ylabel(' ')
plt.xlabel('Time', fontsize=18)
plt.title('Case 2: gamma=0.95', fontsize=20)
plt.legend(loc='center right')
plt.show()
```

Note that in this case $\beta_1=0.71249$ and $\beta_2=0.75$

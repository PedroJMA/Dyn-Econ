** Exercise 1**
Consider the function $f(x) = \frac{1}{1+25x^2}$ defined on $[0,1]$. Approximate $f$ by a Chebyshev polynomial of degree $n$ using $m=n+1$ grid points chosen as the (renormalized) roots of the (n+1)th Chebyshev polynomial. 

Define the absolute error function $h_n(x)$ as 
$$h_n(x) = |f(x) - \hat{f}(x)|.$$

* (a) Plot the function $h_n$ for some values of $n$. 
* (b) Define the relative error function 
$$
r(n) = \sup_{x \in [0,1]} \frac{h_n(x)}{|f(x)|}.
$$
What happens as you increase $n$ ? 
Is there any tendency for $r(n)$ to tend to zero as $n$ tends to infinity? 


```python
from __future__ import division
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

n = 6 #degree of the polynomial 
m = n+1 
points = 100 #number of data points 
xmin = 0
xmax = 1
x = np.linspace(xmin,xmax,points)

#from x in [a,b] to z in [-1,1]
def z(x,a,b):
    z = (2*x - a - b)/(b-a)
    return z

#from z in [-1,1] to y in [a,b]
def y(z,a,b):
    y = ((b-a)*z + a + b)/2 
    return y

z = z(x,xmin,xmax)

#print 'x:',x 
#print 'z:',z

#function to approximate
def f(x):
    f = 1/(1+25*(x**2))
    return f

#print 'f(x)', f(x)
#print 'f(z)', f(z)

#Absolute error function h_n(x) 
def h(x,f,g):
    h = abs(f(x) - g(x))
    return h

#Relative error function 
def r(x,f,g):
    ratio = abs(f(x) - g(x)) / f(x)
    h = max(ratio)
    return h

# Chepyshev polynomial interpolation
p = np.polynomial.Chebyshev.fit(z,f(z),m)


#plt.plot(y(root,xmin,xmax),p(y(root,xmin,xmax)),'o') 
plt.plot(x,f(x),'-',lw=3, label='$f(x)$')
plt.plot(x,p(x),'-',lw=1, label='$\hat{f}(x)$')
plt.title('Chebyshev interpolation of $f=1/(1+25x^2)$' '\n polynomial of degree 6' , fontsize=18)
plt.xlabel('x', fontsize=14)
plt.axhline(y=0,ls='--', lw=.5)
plt.legend(loc=0)
plt.show()

plt.plot(x,f(x)-p(x),'-',lw=2)
#plt.plot(x,0*x,'-',lw=1)
plt.title('Chebyshev polynomial of degree 6' '\n $f(x)$ - $\hat{f}(x)$', fontsize=18)
plt.xlabel('x', fontsize=14)
plt.axhline(y=0,ls='--', lw=.5)
plt.show()

plt.plot(x,f(x),'-',lw=3, label='$f(x)$')
plt.plot(x,h(x,f,p),'-',lw=2, label='$h_6(x)$')
plt.title('Absolute error function', fontsize=18)
plt.xlabel('x', fontsize=14)
plt.legend(loc=0)
plt.show()

#plt.plot(x,h(x,f,p),'-',lw=2)
#plt.plot(x,f(x),'-',lw=3, label='$f(x)$')
plt.plot(x,h(x,f,p)/f(x),'-',lw=2) 
plt.title('Relative error function' '\n $h_6(x)/f(x)$', fontsize=18)
plt.axhline(y=r(x,f,p),ls='--',c='r', lw=1)
plt.show()

print 'Relative error function with n=6 is',r(x,f,p)

#Absolute error function as function of n, given x and f
def aef(n):
    m = n+1
    poly = np.polynomial.Chebyshev.fit(z,f(z),m)
    aef = h(x,f,poly)
    return aef

#Relative error function as function of n, given x and f
def ref(n):
    ratio = aef(n)/f(x)
    ref = max(ratio)
    return ref

plt.plot(x,aef(6),'-.',lw=2.5, label='$n=6$')
plt.plot(x,aef(8),'--',lw=2.5, label='$n=8$')
plt.plot(x,aef(10),'-',lw=2.5, label='$n=10$')
plt.title('Absolute error function' '\n for some values of n', fontsize=18)
plt.xlabel('x', fontsize=14)
plt.legend(loc=0)
plt.show()

#plt.plot(x,h(x,f,p),'-',lw=2)
#plt.plot(x,f(x),'-',lw=3, label='$f(x)$')
plt.plot(x,aef(6)/f(x),'-',lw=1, label='$n=6$')
plt.plot(x,aef(8)/f(x),'-',lw=2, label='$n=8$')
plt.plot(x,aef(10)/f(x),'-',lw=3, label='$n=10$')
plt.title('Relative error function' '\n for some values of n', fontsize=18)
plt.axhline(y=ref(6),ls='-',c='b', lw=1)
plt.axhline(y=ref(8),ls='-',c='g', lw=1)
plt.axhline(y=ref(10),ls='-',c='r', lw=1)
plt.legend(loc=0)
plt.show()

print 'The relative error function with n=6 is',ref(6)
print 'The relative error function with n=8 is',ref(8)
print 'The relative error function with n=10 is',ref(10)

```

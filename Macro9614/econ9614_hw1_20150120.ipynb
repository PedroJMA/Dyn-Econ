### Exercise 1: Simple interpolation

**1**. Consider the function $f(x) = \frac{1}{1 + 25 x^2}$ defined on the interval $[0,1]$. Approximate $f$ by a function 
$$
\hat{f}(x) = a_0 + a_1x + a_2x^2 + \ldots + a_nx^n
$$
choosing the coefficients $a_k$ such that $\hat{f}(x_i) = f(x_i)$ for $x_i = i/n$, $i = 0, 1, \ldots, n$. 
Define the absolute error function $h_n(x)$ via 
$$
h_n(x) = |f(x) - \hat{f}(x)|
$$

**1.(a)** Plot the function $h_n$ for some values of $n$. 

```python
from __future__ import division
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

n = 10 #degree of the polynomial 
x = np.linspace(0,1,n+1)

#function to approximate
def y(x):
    y = 1/(1+25*x**2)
    return y
    
plt.plot(x,y(x),'o') #Function at n+1 nodes
plt.plot(np.arange(0,1,0.01),y(np.arange(0,1,0.01)),'-') #smooth line
plt.title('function $f=1/(1+25x^2)$' '\n with $n+1$ evenly-spaced nodes', fontsize=14)
plt.show()
```

```python
#%matplotlib inline
# Least square polynomial fit that gives us a vector of coefficients p that minimises the squared error.
z=np.polyfit(x,y(x),n)
# A one-dimensional polynomial class; using coefficients from least square polynomial fit
p = np.poly1d(z)

#Here we define the high degree polynomials 
p10= np.poly1d(np.polyfit(x,y(x),10))
p20= np.poly1d(np.polyfit(x,y(x),20))
p30= np.poly1d(np.polyfit(x,y(x),30))

plt.plot(x,y(x),'o') #Function at n+1 nodes
plt.plot(np.arange(0,1,0.01),y(np.arange(0,1,0.01)),'b',lw=2,label='$f$') #smooth function line
plt.plot(np.arange(0,1,0.01),p(np.arange(0,1,0.01)),'g--',lw=2,label='$\hat{f} = p(n)$') #polynomial approx
#plt.plot(np.arange(0,1,0.01),p30(np.arange(0,1,0.01)),'r-',lw=2,label='$\hat{f} = p(30)$') #extreme approx
plt.title('Polynomial interpolation' '\n (monomial basis, evenly-spaced nodes)', fontsize=16)
plt.legend(loc='lower left')
plt.show()
```

```python
#%matplotlib inline
#absolute error function h_n(x) 

def h(x,f,g):
    h = abs(f(x) - g(x))
    return h

hp = h(np.arange(0,1,0.01),y,p)
h10 = h(np.arange(0,1,0.01),y,p10)
h20 = h(np.arange(0,1,0.01),y,p20)
h30 = h(np.arange(0,1,0.01),y,p30)

plt.plot(np.arange(0,1,0.01),hp,'r',lw=2,label='$h_{10(x)}$') 
#plt.plot(np.arange(0,1,0.01),h10,'b-',lw=1,label='$h_{10}(x)$') 
#plt.plot(np.arange(0,1,0.01),h20,'g-',lw=1,label='$h_{20}(x)$') 
#plt.plot(np.arange(0,1,0.01),h30,'k-.',lw=1,label='$h_{30}(x)$') 

plt.legend(loc='upper left')
plt.title('Absolute error function $h_n(x)$', fontsize=16)
plt.show()
```

### Exercise 2: J-period OG environment

**2**. Consider a $J$-period Overlapping Generations environment where individuals maximize
$$
\sum_{j=1}^J \beta^{j-1} \frac{c_j^{1-\sigma}}{1-\sigma}
$$
subject to the following sequence of budget constraints: 
$$
c_j + a_{j+1} = we_j + (1+r)a_j, \quad j = 1, 2, \ldots, J
$$
and the following initial and boundary value conditions: 
$$
a_1 = 0 \quad a_{J+1} = 0. 
$$

Notice that we are considering a stationary environment, i.e. the interest rate $r$ and the wage $w$ are constant and the efficiency units of labour $e_j$ depend only on age and not on cohort. 

In addition, there is no population growth. 
The wage and the interest rate are determined competitively. 



**2(a)** Define a stationary competitive equilibrium

A stationary competitive equilibrium is 
* individual consumption allocations $\{c_j\}_{j=1}^{J}$
* individual asset allocations $\{a_j\}_{j=1}^{J}$
* interest rate $r$, wage $w$
* aggregate labour supply $H$
* aggregate consumption $C$ and 
* capital stock $K$ 

such that 

* Given $r$ and $w$, the allocations $\{c_j\}_{j=1}^{J}$ and $\{a_j\}_{j=1}^{J}$ solve the individual's maximization problem. 
* Given $K$ and $H$, the prices $w$ and $r$ are equal to the marginal products of labour and capital: $w = A(1-\theta)(K/H)^\theta$ and $r = A\theta(H/K)^{1-\theta} - \delta$. 
* Aggregate labour supply is consistent with the endowment profile $\{e_j\}_{j=1}^{J}$ so that $H = \sum_{j=1}^Je_j$.
* Aggregate capital stock is consistent with savings behaviour: $K = \sum_{j=2}^J a_j$.
* Aggregate consumption is consistent with individual consumption behaviour: $C = \sum_{j=2}^J c_j$.
* Aggregate resource constraint satisfies $C + (\delta -1)K = AK^\theta H^{1-\theta}$. 

**2(b)** Solve explicitly for optimal $c_j$ and $a_j$ in terms of $\sigma$, $\beta$, $\{e_j\}_{j=1}^J$, $w$ and $r$. 

The consumption allocation for generation $j$ is 
$$c_j = [\beta(1+r)]^{(j-1)(1/\sigma)}\times c_1 \quad j = 1, \ldots, J$$
where 
$$c_1 = \frac{w \sum_{j=1}^J(1+r)^{J-j}e_j }{ \sum_{j=1}^{J-1} (1+r)^{J-1-j}[\beta(1+r)]^{(j-1)/\sigma} }$$ 
and 
$$a_{j+1} = (1+r)a_j + we_j - c_j$$
that is,
$$a_{j+1} = (1+r)a_j + we_j - [\beta(1+r)]^{(j-1)(1/\sigma)}\times \frac{w \sum_{j=1}^J(1+r)^{J-j}e_j }{ \sum_{j=1}^{J-1} (1+r)^{J-1-j}[\beta(1+r)]^{(j-1)/\sigma} }$$

**2(c)** Sketch an algorith for finding the stationary competitive equilibrium.

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bisection method

def bisect(f, a, b, tol=10e-5):
    lower, upper = a, b
    global point
    point = []
    if upper - lower < tol:
        point1 = 0.5 * (upper + lower)  
        point.append(point1)
        return point
    else:
        middle = 0.5 * (upper + lower)
        if f(middle) > 0:   
            bisect(f, lower, middle)
        else:               
            bisect(f, middle, upper)
        point.append(middle)
        point = point[:1]
    return  point[0]
```

```python
# Parameters
A = 1
J = 12
theta = 0.36 
sigma = 2
delta = 0.35 
ej = [0.36, 0.75, 1.11, 1.32, 1.45, 1.49, 1.47, 1.34, 1.13, 0.61, 0.17, 0]

ky = 0.5 #Capital-Output ratio
```

```python
# interest rate, wage 
r = theta/ky - delta
print 'r =', r
w = (1-theta)/ky
print 'w = ', w
H = sum(ej)
print 'H =', H
```

```python

def c1(beta):
    c1x = w*sum([j*(1+r)**(J-(i+1)) for i,j in enumerate(ej)]) 
    c1y = sum([(1+r)**(J-1-(i+1))*(beta*(1+r))**(i/sigma) for i in range(J-1)])
    c1 = c1x/c1y
    return c1

def cj(beta):
    c2 = [ c1(beta)*(beta*(1+r))**((i+2)/sigma)  for i in range(J-1)]
    consumption = [c1(beta)]
    consumption.extend(c2)
    return consumption

def am(m,sav):
    am = sum([j*(1+r)**(m-(i+1)) for i,j in enumerate(sav[:m])])
    return am

# Allocations 
def allocations(beta):
    global consumption, savings, assets, bigK, bigC, bigS, bigY, bigKY
    c2 = [ c1(beta)*(beta*(1+r))**((i+2)/sigma)  for i in range(J-1)]
    consumption = [c1(beta)]
    consumption.extend(c2)
    savings = [w*i - j for i,j in zip(ej, cj(beta))]
    assets = [am(i+1,savings) for i in range(J-1)]
    bigK = sum(assets)
    bigC = sum(consumption)
    bigS = sum(savings)
    bigY = bigC + bigK #
    bigKY = bigK / bigY
    #return bigK, consumption, savings, assets
    #return bigK, bigC, bigY, bigS, bigKY
    return ky-bigKY
```

```python
#Bisection method
bss = bisect(allocations,0,5)
print 'beta =',bss
```

```python
# plot equilibrium allocations 
%matplotlib inline
import matplotlib.pyplot as plt

grid = np.linspace(1, J,J) #Generations
allocations(bss)
plt.plot(grid, consumption,'-o', linewidth=2, label='with beta at SS')
plt.title('Consumption allocations', fontsize=20)
plt.xlabel('Generation j', fontsize=18)
plt.legend(loc='upper left')
plt.show()
```

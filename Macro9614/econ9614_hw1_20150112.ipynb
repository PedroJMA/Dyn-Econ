### Exercise 1: Simple interpolation

**1**. Consider the function $f(x) = \frac{1}{1 + 25 x^2}$ defined on the interval $[0,1]$. Approximate $f$ by a function 
$$
\hat{f}(x) = a_0 + a_1x + a_2x^2 + \ldots + a_nx^n
$$
choosing the coefficients $a_k$ such that $\hat{f}(x_i) = f(x_i)$ for $x_i = i/n$, $i = 0, 1, \ldots, n$. 
Define the absolute error function $h_n(x)$ via 
$$
h_n(x) = |f(x) - \hat{f}(x)|
$$

**1.(a)** Plot the function $h_n$ for some values of $n$. 

```python
from __future__ import division
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

n = 6 #degree of the polynomial 
x = np.linspace(0,1,n+1)

#function to approximate
def y(x):
    y = 1/(1+25*x**2)
    return y
    
plt.plot(x,y(x),'o') #Function at n+1 nodes
plt.plot(np.arange(0,1,0.01),y(np.arange(0,1,0.01)),'-') #smooth line
plt.title('function $f=1/(1+25x^2)$' '\n with $n+1$ evenly-spaced nodes', fontsize=14)
plt.show()
```

```python
#%matplotlib inline
# Least square polynomial fit that gives us the 
z=np.polyfit(x,y(x),n)
# A one-dimensional polynomial class; 
# using coefficients from least square polynomial fit
p = np.poly1d(z)

#Here we define the high degree polynomials 
p10= np.poly1d(np.polyfit(x,y(x),10))
p20= np.poly1d(np.polyfit(x,y(x),20))
p30= np.poly1d(np.polyfit(x,y(x),30))

plt.plot(x,y(x),'o') #Function at n+1 nodes
plt.plot(np.arange(0,1,0.01),y(np.arange(0,1,0.01)),'b',lw=2,label='$f$') #smooth function line
plt.plot(np.arange(0,1,0.01),p(np.arange(0,1,0.01)),'g--',lw=2,label='$\hat{f} = p(n)$') #polynomial approx
plt.plot(np.arange(0,1,0.01),p30(np.arange(0,1,0.01)),'r-',lw=2,label='$\hat{f} = p(30)$') #extreme approx
plt.title('Polynomial interpolation' '\n (monomial basis, evenly-spaced nodes)', fontsize=16)
plt.legend(loc='lower left')
plt.show()
```

```python
#%matplotlib inline
#absolute error function h_n(x) 

def h(x,f,g):
    h = abs(f(x) - g(x))
    return h

h6 = h(np.arange(0,1,0.01),y,p)
h10 = h(np.arange(0,1,0.01),y,p10)
h20 = h(np.arange(0,1,0.01),y,p20)
h30 = h(np.arange(0,1,0.01),y,p30)

plt.plot(np.arange(0,1,0.01),h6,'r',lw=2,label='$h_6(x)$') 
plt.plot(np.arange(0,1,0.01),h10,'b-',lw=1,label='$h_{10}(x)$') 
plt.plot(np.arange(0,1,0.01),h20,'g-',lw=1,label='$h_{20}(x)$') 
plt.plot(np.arange(0,1,0.01),h30,'k-.',lw=1,label='$h_{30}(x)$') 

plt.legend(loc='upper left')
plt.title('Absolute error function $h_n(x)$', fontsize=16)
plt.show()
```

**1.(b)** Define the relative error funciton 
$$

$$
What happens as you increase $n$ ? Is there any tendency for $r(n)$ to tend to zero as $n$ tends to infinity?

$$
c_j + a_{j+1} = we_j + (1+r)a_j, \quad j = 1, 2, \ldots, J

$$
